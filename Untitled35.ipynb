```python
import numpy as np

def gravitational_wave(M, eta, r, theta, Phi, r_dot, Phi_dot):
    h_plus = -M * eta / r * ((np.cos(2*theta) + 1) * (r_dot**2 + r**2 * Phi_dot**2 + M/r) - 2*r_dot*r*Phi_dot*np.sin(2*Phi))
    h_cross = -2*M * eta / r * np.cos(theta) * (r_dot**2 + r**2 * Phi_dot**2 + M/r) * np.sin(2*Phi) - 2*r_dot*r*Phi_dot*np.cos(2*Phi)
    return h_plus, h_cross

# Example usage:
M = 1.0  # mass
eta = 0.25  # symmetric mass ratio
r = 10.0  # separation
theta = np.pi/4  # inclination angle
Phi = np.pi/2  # orbital phase
r_dot = 0.00000001  # radial velocity
Phi_dot = 0.2  # orbital frequency

h_plus, h_cross = gravitational_wave(M, eta, r, theta, Phi, r_dot, Phi_dot)

print("h_plus:", h_plus)
print("h_cross:", h_cross)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_gravitational_waves(M, eta, r, theta, phi_dot):
    h_plus = -M * eta / r * ((np.cos(2 * theta) + 1) * (r**2 * phi_dot**2 + M / r) - 2 * r * phi_dot * np.sin(2 * theta))
    h_cross = -2 * M * eta / r * np.cos(theta) * (r**2 * phi_dot**2 + M / r) * np.sin(2 * theta) - 2 * r * phi_dot * np.cos(2 * theta)
    return h_plus, h_cross

# Define constants
M = 1  # Mass
eta = 0.0001  # Dimensionless parameter
r = np.linspace(10, 50, 1000)  # Separation
theta = np.pi/4  # Inclination angle
phi_dot = .00000001  # Angular velocity

# Calculate gravitational waves
h_plus, h_cross = calculate_gravitational_waves(M, eta, r, theta, phi_dot)

# Plot Figure 1
plt.plot(r, h_plus, label='Plus-polarized')
plt.plot(r, h_cross, label='Cross-polarized')
plt.xlabel('Separation (r)')
plt.ylabel('Gravitational Wave Strain')
plt.title('Inspiral Strain')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def h_plus(M, eta, r, theta, phi, r_dot, phi_dot):
    return -M * eta / r * ((np.cos(2*theta) + 1) * (r_dot**2 + r**2 * phi_dot**2 + M/r) - 2*r_dot*r_dot*phi_dot*np.sin(2*phi))

def h_cross(M, eta, r, theta, phi, r_dot, phi_dot):
    return -2*M * eta / r * np.cos(theta) * (r_dot**2 + r**2 * phi_dot**2 + M/r) * np.sin(2*phi) - 2*r_dot*r_dot*phi_dot*np.cos(2*phi)

# Define the parameters
M = 1.0  # Mass
eta = 0.25  # Symmetric mass ratio
r = np.linspace(10, 50, 100)  # Distance
theta = np.pi/4  # Inclination angle
phi = 0  # Azimuthal angle
r_dot = 0.1  # Radial velocity
phi_dot = 0.05  # Azimuthal velocity

# Calculate h_plus and h_cross
h_plus_values = h_plus(M, eta, r, theta, phi, r_dot, phi_dot)
h_cross_values = h_cross(M, eta, r, theta, phi, r_dot, phi_dot)

# Plot the results
plt.plot(r, h_plus_values, label='h+')
plt.plot(r, h_cross_values, label='hx')
plt.xlabel('Distance (r)')
plt.ylabel('Strain')
plt.title('Gravitational Wave Strains')
plt.legend()
plt.show()
```
