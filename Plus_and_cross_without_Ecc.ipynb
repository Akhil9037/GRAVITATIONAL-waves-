```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_polarization(theta, phi, strain_amplitude, frequency, time):
    # Calculate the plus and cross polarization components
    plus_polarization = strain_amplitude * (1 + np.cos(theta)**2) * np.cos(2 * np.pi * frequency * time)
    cross_polarization = -2 * strain_amplitude * np.cos(theta) * np.sin(2 * np.pi * frequency * time)

    return plus_polarization, cross_polarization

###### USING THE EQUATION FROM THE PAPER “Gravitational Waves” by Bernard Schutz, published in 2009.



# Define the parameters
theta = np.pi / 4  # Polar angle
phi = np.pi / 3  # Azimuthal angle
strain_amplitude = 0.5  # Strain amplitude
frequency = 10  # Frequency in Hz
time = np.linspace(0, 1, 1000)  # Time array

# Calculate the polarization components
plus, cross = calculate_polarization(theta, phi, strain_amplitude, frequency, time)

# Plot the strain versus time for plus and cross polarization
plt.plot(time, plus, label='Plus Polarization')
plt.plot(time, cross, label='Cross Polarization')
plt.xlabel('Time')
plt.ylabel('Strain')
plt.title('Gravitational Wave Polarizations')
plt.legend()
plt.grid(True)
plt.show()
```
